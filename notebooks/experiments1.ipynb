```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# load the data
def load_data(path : str) -> pd.DataFrame:
    current_dir = os.path.join(os.getcwd())
    parent_dir = os.path.dirname(current_dir)
    data_path = os.path.join(parent_dir, path)

    if not os.path.exists(data_path):
        raise FileNotFoundError(f"Data file not found at {data_path}")
    
    df = pd.read_csv(data_path)
    return df

    
```

```python
# data loading
df = load_data('data/raw/weather_data.csv')
```

```python
# convert time column to datetime with utc timezone 6+
df['time'] = pd.to_datetime(df['time']).dt.tz_convert('Etc/GMT-6')
```

```python
# remove columns those are completely nan values
df = df.dropna(axis=1, how='all')
```
