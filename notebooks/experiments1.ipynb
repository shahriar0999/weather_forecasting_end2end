```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# load the data
def load_data(path : str) -> pd.DataFrame:
    current_dir = os.path.join(os.getcwd())
    parent_dir = os.path.dirname(current_dir)
    data_path = os.path.join(parent_dir, path)

    if not os.path.exists(data_path):
        raise FileNotFoundError(f"Data file not found at {data_path}")
    
    df = pd.read_csv(data_path)
    return df

    
```

```python
# data loading
df = load_data('data/raw/weather_data.csv')
```

```python
# convert time column to datetime with utc timezone 6+
df['time'] = pd.to_datetime(df['time']).dt.tz_convert('Etc/GMT-6')
```

```python
# remove columns those are completely nan values
df = df.dropna(axis=1, how='all')
```

```python
# drop rain column
df = df.drop(columns=['rain'])
```

```python
# drop snowfall column
df = df.drop(columns=['snowfall'])
```

```python
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# Plot each cloud cover histogram
ax[0].hist(df['cloudcover_low'], color='skyblue', alpha=0.7)
ax[0].set_title('Low Cloud Cover')

ax[1].hist(df['cloudcover_mid'], color='orange', alpha=0.7)
ax[1].set_title('Mid Cloud Cover')

ax[2].hist(df['cloudcover_high'], color='green', alpha=0.7)
ax[2].set_title('High Cloud Cover')

# Overall title
fig.suptitle('Cloud Cover Distributions by Level')

plt.tight_layout()
plt.show()
```
