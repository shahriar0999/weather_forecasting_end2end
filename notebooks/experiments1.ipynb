```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# load the data
def load_data(path : str) -> pd.DataFrame:
    current_dir = os.path.join(os.getcwd())
    parent_dir = os.path.dirname(current_dir)
    data_path = os.path.join(parent_dir, path)

    if not os.path.exists(data_path):
        raise FileNotFoundError(f"Data file not found at {data_path}")
    
    df = pd.read_csv(data_path)
    return df

    
```

```python
# data loading
df = load_data('data/raw/weather_data.csv')
```

# Select columns that are need to classify weather condition

```python
features = [
    "time",
    'temperature_2m',
    "relative_humidity_2m",
    "dew_point_2m",
    "apparent_temperature",
    "precipitation",
    "pressure_msl",
    "cloudcover",
    "cloudcover_low",
    "cloudcover_mid",
    "cloudcover_high",
    "windspeed_10m",
    "windgusts_10m",
    "winddirection_10m",
    "sunshine_duration",
    "shortwave_radiation",
    "diffuse_radiation",
    "direct_radiation",
    "terrestrial_radiation",
]

target = df['weather_code'].map(weather_code_map).values

main_df = df[features]

main_df.insert(1, 'weather_code_map', target)
```

```python
features = [
    "temperature_2m", "relative_humidity_2m", "dew_point_2m", "apparent_temperature",
    "precipitation", "pressure_msl", "cloudcover", "windspeed_10m", "sunshine_duration",
    "shortwave_radiation", "direct_radiation", "diffuse_radiation", "windgusts_10m"
]

for feature in features:
    plt.figure(figsize=(10, 4))
    sns.boxplot(data=main_df, x='weather_code_map', y=feature)
    plt.title(f'{feature} vs Weather Code')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
for feature in ["temperature_2m", "precipitation", "cloudcover"]:
    plt.figure(figsize=(10, 5))
    for label in main_df['weather_code_map'].unique():
        subset = main_df[main_df['weather_code_map'] == label]
        sns.kdeplot(subset[feature], label=f"Weather Code {label}", fill=True)
    plt.title(f"KDE of {feature} by Weather Condition")
    plt.legend()
    plt.show()
```

```python
from scipy.stats import f_oneway

for feature in features:
    groups = [main_df[main_df['weather_code_map'] == label][feature] for label in main_df['weather_code_map'].unique()]
    stat, p = f_oneway(*groups)
    print(f"{feature}: F={stat:.2f}, p-value={p:.4f}")
```

```python
from sklearn.feature_selection import mutual_info_classif

X = main_df[features]
y = main_df['weather_code_map']
mi_scores = mutual_info_classif(X, y, discrete_features=False)

mi_df = pd.DataFrame({'Feature': features, 'MI Score': mi_scores})
mi_df = mi_df.sort_values(by='MI Score', ascending=False)

sns.barplot(data=mi_df, x='MI Score', y='Feature')
plt.title("Mutual Information with Weather Condition")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

X = main_df[features]
y = LabelEncoder().fit_transform(main_df['weather_code_map'])

model = RandomForestClassifier()
model.fit(X, y)

importances = model.feature_importances_

fi_df = pd.DataFrame({'Feature': features, 'Importance': importances})
fi_df = fi_df.sort_values(by='Importance', ascending=False)

sns.barplot(data=fi_df, x='Importance', y='Feature')
plt.title("Random Forest Feature Importance")
plt.show()
```
